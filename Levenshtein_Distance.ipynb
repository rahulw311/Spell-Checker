```python
#import of all required libraries
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import string
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from urllib import request
import re
import random
from collections import defaultdict
```

```python
#function to preprocess the input text file
def text_preprocessing(text):
    # convert text to lowercase
    text = text.lower()
    #print(text)

    #word tokenizing
    tokens = word_tokenize(text)

    # removing numbers, stopwords, and punctuation
    lang_stopwords = stopwords.words("english")
    tokens = [token for token in tokens if not token.isdigit() and \
                            not token in string.punctuation and \
                                token not in lang_stopwords]

    # join tokens and form string
    preprocessed_text = " ".join(tokens)

    return preprocessed_text

# reading the input text file and converting it into a single string
corpus=""
with open('/content/drive/MyDrive/cs173/hw1/MobyDick.txt') as f:
    corpus=corpus.join(f.readlines())

# no of sentences
tokens = sent_tokenize(corpus)
print("No of sentences=",len(tokens))

#no of tokens
tokens = word_tokenize(corpus)
print("No of tokens=",len(tokens))

#no of words
print("No of words using split=",len(corpus.split()))

#function call on the created string to preprocess the text
mod_corpus = text_preprocessing(corpus)
print(mod_corpus)
```

```python
#forming a list of words in the corpus
words=mod_corpus.split(" ")

#formualtion of a dictionary of unique words in the corpus
x=defaultdict(str)
for word in words:
  x[word]=1
```

```python
#storing all the unique words in a list
key=list(x.keys())
```

```python
#downlloading a dictionary from github
dicturl="https://raw.githubusercontent.com/dwyl/english-words/master/words_alpha.txt"
request.urlretrieve(dicturl,"word_dict")
```

```python
#creating a list of all words in the dictionary
word_dict=[]
f=open("/content/word_dict","r")
for line in f:
  word=line.strip()
  word=word.translate(str.maketrans('', '', string.punctuation))
  word_dict.append(word)
f.close()
```

```python
#finding list of misspelt words
misspelt=[]
for word in key:
  #eliminating numbers from words
  if any(c.isdigit() for c in word):
    continue
  #eliminating hyphenated words
  if any(c in string.punctuation for c in word):
    continue
  #eliminating all roman numerals
  pattern = "^[ivxlcdm]*$"
  x = re.match(pattern, word)
  if(x):
    continue
  #finding spelling errors
  if word not in word_dict:
    print(word)
    misspelt.append(word)
```

```python
#printing number of misspelt words in the entire corpus
print(len(misspelt))
```

```python
#implementation of edit distance using levenshtein algorithm
def levenshtein(a, b):
    if not a: return len(b)
    if not b: return len(a)
    return min(levenshtein(a[1:], b[1:])+(a[0] != b[0]),
               levenshtein(a[1:], b)+1,
               levenshtein(a, b[1:])+1)
```

```python
#generating suggestions based on the spelling errors found
for i in misspelt:
  #You can alter the length of the misspelt words taken into consideration here
  if(len(i)>4):
    continue
  temp=[]
  suggestions=[]
  for j in word_dict:
    if(len(i)<=len(j)+1 and len(i)>=len(j)-1):
        temp.append(j)
  #print(temp)
  #print(len(temp))
  for word in temp:
    if(levenshtein(i,word)<=2):
      suggestions.append(word)
  if(len(suggestions)<=3):
    print(i,":",suggestions)
  else:
    print(i,":",random.choices(suggestions,k=3))
```
